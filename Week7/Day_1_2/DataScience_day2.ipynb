```python
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plot

trials = 5000
perc = 80
x = np.random.normal(170,7,trials)
ar = []
p = np.percentile(x, perc)

for n in x:
    if n > p:
        ar.append(n)
        if p >= len(ar):
            break
ar
l = np.where(np.logical_and(x>=ar[0], x<=10000))
len(l[0])
```
